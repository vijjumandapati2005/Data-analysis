```python
import numpy as np
import matplotlib.pyplot as plt

# Define the linear function
def linear_function(x):
    return 2 * x + 3

# Generate some sample data
x = np.linspace(0, 10, 100)
y = linear_function(x)

# Generate some random points above and below the line
np.random.seed(0)
random_points = np.random.rand(20) * 10
random_offsets = (np.random.rand(20) - 0.5) * 10
x_above = random_points[random_offsets > 0]
y_above = linear_function(x_above) + random_offsets[random_offsets > 0]
x_below = random_points[random_offsets <= 0]
y_below = linear_function(x_below) + random_offsets[random_offsets <= 0]

# Plot the linear function
plt.plot(x, y, color='black', label='Linear Function')

# Plot points above and below the line with different symbols
plt.scatter(x_above, y_above, color='blue', marker='^', label='Above Line')
plt.scatter(x_below, y_below, color='green', marker='o', label='Below Line')

# Add labels and legend
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Function with Points Above and Below the Line')
plt.legend()

# Show plot
plt.grid(True)
plt.show()
```
